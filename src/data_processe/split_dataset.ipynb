```python
import pandas as pd


relation_filepath = "/Users/xuhaoshuai/Project/HumanIE-IPM-experiment-2.0/relation_data/parentschild/"

all_data_df = pd.read_csv(relation_filepath + "all_data_deduplication.csv", index_col='sent_id')
all_data_df
```

分割训练集

```python
sample_pos = all_data_df[all_data_df['label'] == 1].sample(n=500)
sample_neg = all_data_df[all_data_df['label'] == -1].sample(n=500)

labeled_set = pd.concat([sample_pos, sample_neg])
labeled_set = labeled_set.rename(columns={'label': 'human'})
labeled_set.to_csv(relation_filepath + "1_sent_to_mine/1.csv")
```

分割测试集和验证集

```python
data_left = all_data_df[~all_data_df.index.isin(labeled_set.index)]
data_left
```

```python
valid_df = validation_set = data_left.sample(frac=0.3)
valid_df
```

```python
test_df = data_left[~data_left.index.isin(valid_df.index)]
test_df
```

```python
test_df.to_csv(relation_filepath + "4_sent_to_match/1.csv")
valid_df.to_csv(relation_filepath + "validation/validation_data.csv")
```
